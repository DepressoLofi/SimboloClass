```python
trainX = test.iloc[:, 0:-1]
trainY = test.iloc[:, -1]
```

```python
testX = test.iloc[:, 0:-1]
testY = test.iloc[:, -1]
```

## 4. Model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(trainX, trainY)
```

## 5. Evaluate

```python
y_pred = model.predict(testX)
```

```python
y_pred
```

```python
testY
```

```python
mse = np.mean((testY - y_pred)**2)
```

```python
mse
```

```python
total = 0
for i, j in zip(testY, y_pred):
  total += abs(i - j)

print("MAE", total/len(y_pred))
```

```python
total = 0
for i, j in zip(testY, y_pred):
  total += (i - j) ** 2

print("MSE", total/len(y_pred))
```
